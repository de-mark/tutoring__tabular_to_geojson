```python
import pandas as pd
import json
```

```python
full_df = pd.read_csv("data/clean/combined.csv")
full_df
```

```python
geojson = {
       "type": "FeatureCollection",
       "features": []
       }
geojson
```

```python
full_df.columns
```

```python
def create_feature(row):
  res = {
  "type": "Feature",
  "geometry": {
    "type": "Point",
    "coordinates": [row["Longitude (average)"], row["Latitude (average)"]]
  },
  "properties": {
    "country_name":row["country"],
    "country_name_abbr":row["country_abbr"],
    "CO2_1990":row["CO2_1990"],
    "CO2_2005": row["CO2_2005"],
    "CO2_2017":row["CO2_2017"],
    "CO2_2017_per_land_area":row["CO2_2017_per_land_area"],
    "CO2_2017_per_capita":row["CO2_2017_per_capita"]
  }
  }
  return res

for i in full_df.index:
  res = create_feature(full_df.iloc[i])
  geojson["features"].append(res)

geojson
```

```python
with open('data/clean/emissions.json', 'w') as output_file:
    json.dump(geojson, output_file)
```
